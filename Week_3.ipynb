```python
import pandas as pd
```

**Importing the data from wikipedia using panda's read_HTML method**

```python
df_3 = df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

```python
df_3
```

**Cleaning the data, separating the info from one cell into multiple cells based on position or a '(' delimiter**

```python
df_stack = df_3.stack().reset_index()
```

```python
df_stack_2 = df_stack.drop(axis = 1, columns = ['level_0', 'level_1'])
```

```python
df_stack_2.rename(columns = {0:'Column_1'}, inplace = True)
```

```python
df_stack_3 = df_stack_2[~(df_stack_2['Column_1'].str.contains("Not assigned"))]
```

```python
df_stack_3['Postcode'] = df_stack_3['Column_1'].str[:3]
```

```python
df_stack_3['Column_1'] = df_stack_3['Column_1'].str[3:]
```

```python
new = df_stack_3['Column_1'].str.split("(", n = 1, expand = True)
```

```python
new[1] = new[1].str.slice(0,-1)
```

**Further cleaning, separating the columns into multiple values**

```python
new
```

```python
df_stack_3['Borough'] = new[0]
df_stack_3['Neighborhood'] = new[1]
```

```python
df_tor = df_stack_3[['Postcode', 'Borough', 'Neighborhood']]
```
